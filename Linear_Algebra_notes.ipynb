### For a matrix (or the coefficient matrix of a linear equation system), we have the following relationshiop:
$$\#\_of\_columns=dim(Columns\_Space)+dim(Null\_Space)$$

$\vec{0}$ is linearly dependent, since there exist a non-trivial linear combination satisfying $\vec{0}=1$

### For the linear equation system $A \cdot \vec{x}=\vec{b}$
__1. Is there any solution?__<br />
If $\vec{b}$ can be expressed as a linear conbination of the column vectors of $A$, there is one or more solution. Ohterwise, no solution. (If $\vec{b}$ is in the space spanned by the column vectors of $A$).
   
__2. Is the solution unique? (Assume the system is solvable)__<br />
If all the column vectors of $A$ are linearly independent, the solution is unique. Otherwise, more than one solutions.


>If the column vectors of $A$ is linearly independent, we have $\#\_of\_columns=dim(Columns\_Space)$. So $dim(Null\_Space)=0$. This conresponds the case of unique solution.

>If the column vectors of $A$ is linearly dependent, $dim(Null\_Space)\ne 0$. We can add any vector in Null Space to the solution, resulting is a new solution to the system. That's why the solution is no unique.

>All the solutions for $A \cdot \vec{x}=\vec{b}$ consist of a unique solution for $A \cdot \vec{x}=\vec{b}$ plus the Null Space of the matrix $A$ (i.e. the general solution for $A \cdot \vec{x}=\vec{0}$).

__An Example of extreme cases:__ <br \>
$0 \cdot x = 0$ ---Solvable. Infinite solutions. <br \>
$3 \cdot x = 0$ ---Solvable. Unique solution. <br />
$3 \cdot x = 3$ ---Solvable. Unique solution. <br />
$0 \cdot x = 3$ ---Solvable. Unsolvable.

### Property of Gaussian Elimination
- Gaussian elimination doesn't change the Null Space
- Gaussian elimination will change the Column Space

For Example:

Consider this matrix
$$A =
\begin{bmatrix}
    10 & 2 & 14 & 20 & 8 \\
    3 & 5 & 23 & 6 & -2 \\
    13 & 7 & 27 & 26 & 6 \\ 
\end{bmatrix} \qquad
A\_Column\_Space=
\begin{bmatrix}
    a \\
    b \\
    a+b \\ 
\end{bmatrix}
$$
<br />
After Guassian elimination, the matrix is
$$ A\_G\_Elim = 
\begin{bmatrix}
    1 & 0 & 1 & 2 & 1 \\
    0 & 1 & 2 & 0 & -1 \\
    0 & 0 & 0 & 0 & 0 
\end{bmatrix}\qquad
Column\_Space = 
\begin{bmatrix}
    a \\
    b \\
    0 \\ 
\end{bmatrix}
$$

The column space is changed

If we have $A \cdot B = C$ ($A$, $B$ and $C$ are matrices), we have the following relationship:
$$R_A \supseteq R_C$$
$$N_B \subseteq N_C$$
i.e. <br \>
The column space of $C$ is a subspace of the column space of $A$. <br \>
The null space of $B$ is a subspace of the null space of $C$.

#### Invere and Transpose

$(A \cdot B)^{-1} = B^{-1} \cdot A^{-1}$ (if $A$ and $B$ are invertable) <br \>
$(A \cdot B)^{T} = B^{T} \cdot A^{T}$ <br \>
$A^{T} \cdot A$ is symetric: $(A^{T} \cdot A)^{T} = A^{T} \cdot (A^{T})^{T} = A^{T} \cdot A$

### LU decomposition
To do a LU decompostion, we just record the Gaussian elimination procedure useing elementary matrices. 

Example: Consider the matrix
$
A=
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 10 \\
\end{bmatrix}
$,
We want to get $A=LU$
<br />
<br />
$
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & -2 & 1 \\
\end{bmatrix}
\cdot
\begin{bmatrix}
1 & 0 & 0 \\
-4 & 1 & 0 \\
-7 & 0 & 1 \\
\end{bmatrix}
\cdot
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 10 \\
\end{bmatrix}
=
\begin{bmatrix}
1 & 2 & 3 \\
0 & -3 & -6 \\
0 & 0 & 1 \\
\end{bmatrix}
$
<br />
<br />
Then 
<br />
$
U=
\begin{bmatrix}
1 & 2 & 3 \\
0 & -3 & -6 \\
0 & 0 & 1 \\
\end{bmatrix} \qquad \quad
L=
\Bigg(
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & -2 & 1 \\
\end{bmatrix}
\cdot
\begin{bmatrix}
1 & 0 & 0 \\
-4 & 1 & 0 \\
-7 & 0 & 1 \\
\end{bmatrix}
\Bigg)^{-1}
=
\Bigg(
\begin{bmatrix}
1 & 0 & 0 \\
-4 & 1 & 0 \\
-7 & 0 & 1 \\
\end{bmatrix}
\Bigg)^{-1}
\cdot
\Bigg(
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & -2 & 1 \\
\end{bmatrix}
\Bigg)^{-1}
=
\begin{bmatrix}
1 & 0 & 0 \\
4 & 1 & 0 \\
7 & 2 & 1 \\
\end{bmatrix}
$

### Why Eigenvectors Are Important
Lemma: Eigenvectors corresponding different eigenvalues are linearly independent.

Assume $\vec{e_1}$, $\vec{e_2}$ and $\vec{e_3}$ are the eigenvectors of matrix A, corresponding to the eigenvalues $\lambda_1$, $\lambda_2$ and $\lambda_3$. $\vec{v}$ is a vetor. If $\vec{v}$ can be expressed as the linear combination of $A$'s eigenvactors:
$$\vec{v} = \alpha_1 \cdot \vec{e_1} + \alpha_2 \cdot \vec{e_2} + \alpha_3 \cdot \vec{e_3}$$
Then if we do the linear transpose representing by $A$ on the vector $\vec{v}$, it would be bery easy to find the new coordinates:
\begin{align} 
A \cdot \vec{v} & = \alpha_1 \cdot A \cdot \vec{e_1} + \alpha_2 \cdot A \cdot \vec{e_2} + \alpha_3 \cdot A \cdot \vec{e_3} \\
& = \alpha_1 \cdot \lambda_1 \cdot \vec{e_1} + \alpha_2 \cdot \lambda_2 \cdot \vec{e_2} + \alpha_3 \cdot \lambda_3 \cdot \vec{e_3}  
\end{align}
<br>
This is a reason why eigenvector is so important.

### Eigenvalue Algorithm
Derivation:
\begin{align}
& A \cdot \vec{x} = \lambda \cdot \vec{x} \\
\Rightarrow \quad & A \cdot \vec{x} - \lambda \cdot \vec{x} = \vec{0} \\
\Rightarrow \quad & A \cdot \vec{x} - \lambda \cdot I \cdot \vec{x} = \vec{0} \\
\Rightarrow \quad & (A - \lambda \cdot I) \cdot \vec{x} = \vec{0} 
\end{align}

Our goal: <br>
Find appropriate $\lambda$ such that the previous system has non-trivial solution. <br>
$\Leftrightarrow$ The matrix $(A - \lambda \cdot I)$ has non-trivial null space. <br>
$\Leftrightarrow$ The matrix $(A - \lambda \cdot I)$ is singular. <br>
$\Leftrightarrow$ The deternimant $|A - \lambda \cdot I|=0$ . <br>

#### Properties of Eigenvalues
The sum of the trace equals the sum of the eigenvalues. <br>
The product of the trace equals the determinant of the matrix. <br> 
$|A| = |A^T| \Leftrightarrow$ Transpose doesn't change eigenvalues.
